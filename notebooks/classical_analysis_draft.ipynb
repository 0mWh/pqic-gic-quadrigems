# Tree Pruning

```python
import bct  #pip install bctpy
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import networkx as nx
```

```python
# Create thresholds
thr_07 = bct.threshold_absolute(corrs.to_numpy(), 0.7)
thr_08 = bct.threshold_absolute(corrs.to_numpy(), 0.8)
thr_09 = bct.threshold_absolute(corrs.to_numpy(), 0.9)

# Extract tuning curves and coordinates
tuning_curves_sig = tuning_curves.loc[neurons_sig].to_numpy()
coords_sig = coords.loc[neurons_sig].to_numpy()

# Compute signal correlation matrix from extractions
n_neurons = tuning_curves_sig.shape[0]
sc_matrix = np.zeros((n_neurons, n_neurons))

for i in range(n_neurons):
    for j in range(i, n_neurons):
        r, _ = pearsonr(tuning_curves_sig[i], tuning_curves_sig[j])
        sc_matrix[i, j] = r
        sc_matrix[j, i] = r
```

```python
# Full Network Graph
G_full = nx.Graph()
for i in range(n_neurons):
    G_full.add_node(i, pos=coords_sig[i])

for i in range(n_neurons):
    for j in range(i + 1, n_neurons):
        if not np.isnan(sc_matrix[i, j]):
            G_full.add_edge(i, j, weight=sc_matrix[i, j])

# O.7 Threshold Graph
threshold = 0.7
G_thresh_7 = nx.Graph()
for i in range(n_neurons):
    G_thresh_7.add_node(i, pos=coords_sig[i])

for i in range(n_neurons):
    for j in range(i + 1, n_neurons):
        if sc_matrix[i, j] >= threshold:
            G_thresh_7.add_edge(i, j, weight=sc_matrix[i, j])

# O.8 Threshold Graph
threshold = 0.8
G_thresh_8 = nx.Graph()
for i in range(n_neurons):
    G_thresh_8.add_node(i, pos=coords_sig[i])

for i in range(n_neurons):
    for j in range(i + 1, n_neurons):
        if sc_matrix[i, j] >= threshold:
            G_thresh_8.add_edge(i, j, weight=sc_matrix[i, j])

# O.9 Threshold Graph
threshold = 0.9
G_thresh_9 = nx.Graph()
for i in range(n_neurons):
    G_thresh_9.add_node(i, pos=coords_sig[i])

for i in range(n_neurons):
    for j in range(i + 1, n_neurons):
        if sc_matrix[i, j] >= threshold:
            G_thresh_9.add_edge(i, j, weight=sc_matrix[i, j])

# Minimum Spanning Tree (MST)
dist_matrix = 1 - sc_matrix
G_dist = nx.Graph()
for i in range(n_neurons):
    G_dist.add_node(i, pos=coords_sig[i])

for i in range(n_neurons):
    for j in range(i + 1, n_neurons):
        if not np.isnan(dist_matrix[i, j]):
            G_dist.add_edge(i, j, weight=dist_matrix[i, j])

G_mst = nx.minimum_spanning_tree(G_dist)
```

```python
# Network Graphs by Threshold & MST
def plot_graph_3d(G, title, ax, node_color='royalblue'):
    pos = nx.get_node_attributes(G, 'pos')
    weights = np.array([G[u][v]['weight'] for u, v in G.edges()])
    cmap = plt.cm.viridis
    norm = plt.Normalize(vmin=weights.min(), vmax=weights.max())

    for u, v in G.edges():
        x = [pos[u][0], pos[v][0]]
        y = [pos[u][1], pos[v][1]]
        z = [pos[u][2], pos[v][2]]
        weight = G[u][v]['weight']
        edge_color = cmap(norm(weight))
        ax.plot(x, y, z, color=edge_color, alpha=0.6)

    xs, ys, zs = zip(*pos.values())
    ax.scatter(xs, ys, zs, color=node_color, s=10)

    ax.set_title(title, fontsize=10)
    ax.set_axis_off()

fig = plt.figure(figsize=(16, 10))

graphs = [
    (G_full, "Full Network"),
    (G_thresh_7, "Threshold ≥ 0.7"),
    (G_thresh_8, "Threshold ≥ 0.8"),
    (G_thresh_9, "Threshold ≥ 0.9"),
    (G_mst, "Minimum Spanning Tree (MST)")
]

for i, (G, title) in enumerate(graphs):
    ax = fig.add_subplot(2, 3, i + 1, projection='3d')
    plot_graph_3d(G, title, ax)

plt.tight_layout()
plt.show()
```

```python
 # Circular Version - Network Graphs & MST
def plot_graph_circular(G, title, ax):
    pos = nx.circular_layout(G)
    nx.draw_networkx_nodes(G, pos, ax=ax, node_size=10, node_color='royalblue')
    weights = np.array([G[u][v]['weight'] for u, v in G.edges()])
    cmap = plt.cm.viridis
    norm = plt.Normalize(vmin=weights.min(), vmax=weights.max())
    edge_colors = [cmap(norm(G[u][v]['weight'])) for u, v in G.edges()]
    nx.draw_networkx_edges(G, pos, ax=ax, edge_color=edge_colors, alpha=0.6)

    ax.set_title(title, fontsize=10)
    ax.set_axis_off()

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

graphs = [
    (G_full, "Full Network (Circular)"),
    (G_thresh_7, "Threshold ≥ 0.7 (Circular)"),
    (G_thresh_8, "Threshold ≥ 0.8 (Circular)"),
    (G_thresh_9, "Threshold ≥ 0.9 (Circular)"),
    (G_mst, "Minimum Spanning Tree (Circular)")
]

for i, (G, title) in enumerate(graphs):
    plot_graph_circular(G, title, axes[i])

# Hide the last unused subplot
axes[-1].axis('off')

plt.tight_layout()
plt.show()
```

# Rentian Scaling

```python
from scipy import stats
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import networkx as nx
import numpy as np
from sklearn.metrics import r2_score
from scipy.optimize import curve_fit
```

```python
# Rentian Scaling Function

# Notes: this is an updated version from bctpy
def my_rentian_scaling(A, xyz, n, seed=42):
    rng = np.random.default_rng(seed)
    m = np.size(xyz, axis=0)  # find number of nodes in system

    # rescale coordinates so they are all greater than unity
    xyzn = xyz - np.tile(np.min(xyz, axis=0) - 1, (m, 1))

    # find the absolute minimum and maximum over all directions
    nmax = np.max(xyzn)
    nmin = np.min(xyzn)

    count = 0
    N = np.zeros((n,))
    E = np.zeros((n,))

    # create partitions and count the number of nodes inside the partition (n)
    # and the number of edges traversing the boundary of the partition (e)
    while count < n:
        # define cube endpoints
        randx = np.sort((nmax - nmin) * rng.random((2,)))
        randy = np.sort((nmax - nmin) * rng.random((2,)))
        randz = np.sort((nmax - nmin) * rng.random((2,)))

        # ensure cube is within network boundaries
        randx[0] = max(randx[0], nmin)
        randx[1] = min(randx[1], nmax)
        randy[0] = max(randy[0], nmin)
        randy[1] = min(randy[1], nmax)
        randz[0] = max(randz[0], nmin)
        randz[1] = min(randz[1], nmax)

        # find nodes in cube
        l1 = xyzn[:, 0] > randx[0]
        l2 = xyzn[:, 0] < randx[1]
        l3 = xyzn[:, 1] > randx[0]
        l4 = xyzn[:, 1] < randx[1]
        l5 = xyzn[:, 2] > randx[0]
        l6 = xyzn[:, 2] < randx[1]

        L, = np.where((l1 & l2 & l3 & l4 & l5 & l6).flatten())
        if np.size(L):
            # count edges crossing at the boundary of the cube
            E[count] = np.sum(A[np.ix_(L, np.setdiff1d(range(m), L))])
            # count nodes inside of the cube
            N[count] = np.size(L)
            count += 1

    return N, E
```

```python
# Plot of MST Network with Rentian Scaling
def plot_3d_graph(G, title="3D Functional Network", node_size=20, edge_color='gray'):
    fig = plt.figure(figsize=(7, 6))
    ax = fig.add_subplot(111, projection='3d')

    pos = nx.get_node_attributes(G, 'pos')
    xs = [pos[i][0] for i in G.nodes()]
    ys = [pos[i][1] for i in G.nodes()]
    zs = [pos[i][2] for i in G.nodes()]

    ax.scatter(xs, ys, zs, color='royalblue', s=node_size)
    for u, v in G.edges():
        x = [pos[u][0], pos[v][0]]
        y = [pos[u][1], pos[v][1]]
        z = [pos[u][2], pos[v][2]]
        ax.plot(x, y, z, color=edge_color, alpha=0.6, linewidth=1)

    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    plt.tight_layout()
    plt.show()

for i, coord in enumerate(coords_sig):
    G_mst.nodes[i]['pos'] = coord

plot_3d_graph(G_mst, title="MST Functional Network (3D)")
```

```python
# Run Rentian Scaling Function
threshold = 0.9
adj_matrix = (sc_matrix >= threshold).astype(int)
np.fill_diagonal(adj_matrix, 0)

N, E = my_rentian_scaling(adj_matrix, coords_sig, n=300)
mask = (N > 0) & (E > 0)
log_N = np.log10(N[mask]).reshape(-1, 1)
log_E = np.log10(E[mask]).reshape(-1, 1)

model = LinearRegression().fit(log_N, log_E)
slope = model.coef_[0][0]
intercept = model.intercept_[0]
r2 = model.score(log_N, log_E)

log_N_fit = np.linspace(log_N.min(), log_N.max(), 100).reshape(-1, 1)
log_E_fit = model.predict(log_N_fit)
```

```python
# Rentian Scaling with Linear Regression
def rent_exp_linreg(A, coords, M=20):
    # M = 1000  # number of partitions, if we have at least M more than neurons, then each partition has about 1 neuron

    N, E = my_rentian_scaling(A, coords.to_numpy(), M, seed=42)
    # bctpy docs say recommended step is to prune partitions with N < M/2, but according to matlab function header, setting random boxes
    N_prime = N # [N < M/2]
    E_prime = E # [N < M/2]

    # Log-log transformation
    log_N_prime = np.log10(N_prime)
    log_E_prime = np.log10(E_prime)

    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x=log_N_prime, y=log_E_prime)

    # The Rent's exponent is the slope
    rent_exponent = slope
    print("Rent's exponent:", rent_exponent)

    # Standard error of the estimation
    std_error = std_err
    print("Standard error of the estimation:", std_error)
    return rent_exponent, std_error



# Rentian Scaling Exponential SM
def rent_exp_sm(A, coords, M=20):
    # M = 1000  # number of partitions, if we have at least M more than neurons, then each partition has about 1 neuron

    N, E = my_rentian_scaling(A, coords.to_numpy(), M, seed=42)
    # recommended step is to prune partitions with N < M/2
    N_prime = N # [N < M/2]
    E_prime = E # [N < M/2]

    # Log-log transformation
    log_N_prime = np.log10(N_prime)
    log_E_prime = np.log10(E_prime)

    # Add a constant (intercept) to the independent value
    log_N_prime = sm.add_constant(log_N_prime)

    # Fit the robust linear model
    model = sm.RLM(log_E_prime, log_N_prime, M=statsmodels.robust.norms.TukeyBiweight()) # , M=sm.robust.norms.HuberT())
    results = model.fit()

    # The Rent's exponent is the coefficient of the slope
    rent_exponent = results.params[1]
    print("Rent's exponent:", rent_exponent)

    # Standard error of the estimation
    std_error = results.bse[1]
    print("Standard error of the estimation:", std_error)
    return rent_exponent, std_error



# Rentian Scaling Exponential Theilslopes
def rent_exp_theilslopes(A, coords, M=20):
    # M = 1000  # number of partitions, if we have at least M more than neurons, then each partition has about 1 neuron

    N, E = my_rentian_scaling(A, coords.to_numpy(), M, seed=42)
    # recommended step is to prune partitions with N < M/2
    N_prime = N # [N < M/2]
    E_prime = E # [N < M/2]

    # Log-log transformation
    log_N_prime = np.log10(N_prime)
    log_E_prime = np.log10(E_prime)

    slope, intercept, low_slope, high_slope = stats.theilslopes(y=log_E_prime, x=log_N_prime)
    # The Rent's exponent is the coefficient of the slope
    print("Rent's exponent:", slope)

    # Standard error of the estimation
    print("Low and high of exp:", low_slope, high_slope)
    return slope, (low_slope, high_slope)
```

```python
# Rentian Scaling Power Law

# Define the power law function
def power_law(x, a, b):
    return a * x**b

# Define the exponential law function
def exponential_law(x, a, b):
    return a * np.exp(b * x)

N, E = my_rentian_scaling(thr_09, coords.to_numpy(), 20)
X, Y = N, E

# Fit the power law
popt_power, pcov_power = curve_fit(power_law, X, Y)
y_fit_power = power_law(X, *popt_power)

# Fit the exponential law
popt_exponential, pcov_exponential = curve_fit(exponential_law, X, Y)
y_fit_exponential = exponential_law(X, *popt_exponential)

def r2(y, y_fit):
    ss_res = np.sum((y - y_fit)**2)
    ss_tot = np.sum((y - np.mean(y))**2)
    return max(0, 1 - (ss_res / ss_tot))

# Calculate the R^2 values
r2_power = r2_score(Y, y_fit_power)
r2_exponential = r2_score(Y, y_fit_exponential)

print("Power law R^2: {:.2f}".format(r2_power))
print("Exponential law R^2: {:.2f}".format(r2_exponential))

# If power law fit > exponential law fit, then rentian scaling is achieved
```

```python
def compute_small_world_propensity(G, n_rand=100, seed=42): # Increased n_rand
    rng = np.random.default_rng(seed)

    # Work with the largest connected component if the graph is not connected
    if not nx.is_connected(G):
        print("Graph is disconnected. Analyzing the largest connected component.")
        G = G.subgraph(max(nx.connected_components(G), key=len)).copy()
        if G.number_of_nodes() == 0:
            print("Largest connected component is empty. Cannot compute SWP.")
            return np.nan, {}


    # Observed graph metrics
    C_obs = nx.average_clustering(G)
    try:
        L_obs = nx.average_shortest_path_length(G)
    except nx.NetworkXNoPath:
         print("No path exists between some nodes in the observed graph. SWP may be invalid.")
         L_obs = np.nan # Or np.inf, depending on desired behavior
    except Exception as e:
        print(f"An error occurred calculating L_obs: {e}")
        L_obs = np.nan # Generic error handling


    # Create random graphs with same degree sequence
    degrees = [d for _, d in G.degree()]
    rand_Cs = []
    rand_Ls = []

    for _ in range(n_rand):
        G_rand = nx.configuration_model(degrees, seed=rng)
        G_rand = nx.Graph(G_rand)  # remove multiedges
        G_rand.remove_edges_from(nx.selfloop_edges(G_rand))

        # Analyze the largest connected component of the random graph
        if not nx.is_connected(G_rand):
            G_rand = G_rand.subgraph(max(nx.connected_components(G_rand), key=len)).copy()

        if G_rand.number_of_nodes() > 0:
            rand_Cs.append(nx.average_clustering(G_rand))
            try:
                rand_Ls.append(nx.average_shortest_path_length(G_rand))
            except nx.NetworkXNoPath:
                rand_Ls.append(np.nan) # Append nan for disconnected random graphs
            except Exception as e:
                 print(f"An error occurred calculating L_rand for a random graph: {e}")
                 rand_Ls.append(np.nan) # Generic error handling

    C_rand = np.nanmean(rand_Cs)
    L_rand = np.nanmean(rand_Ls)

    # Create lattice graph with similar size
    n = G.number_of_nodes()
    k = int(np.mean(degrees))
    # Ensure k is valid for watts_strogatz_graph (k < n)
    k = min(k, n - 1) if n > 1 else 0
    if k == 0 and n > 1: # Handle case where mean degree is 0 but there are nodes
         print("Mean degree is 0. Creating an empty lattice graph.")
         G_latt = nx.Graph() # Empty graph
    elif n <= 1: # Handle case with 0 or 1 node
         print("Graph has 0 or 1 node. Creating an empty lattice graph.")
         G_latt = nx.Graph()
    else:
        G_latt = nx.watts_strogatz_graph(n, k, 0, seed=rng)

    C_latt = nx.average_clustering(G_latt) if G_latt.number_of_nodes() > 1 else 0
    try:
        if G_latt.number_of_nodes() > 1 and nx.is_connected(G_latt):
            L_latt = nx.average_shortest_path_length(G_latt)
        elif G_latt.number_of_nodes() <= 1:
            L_latt = 0 # Path length is 0 for single node or empty graph
        else:
            print("Lattice graph is disconnected; assigning nan path length.")
            L_latt = np.nan
    except nx.NetworkXNoPath:
         print("No path exists in the lattice graph. SWP may be invalid.")
         L_latt = np.nan
    except Exception as e:
        print(f"An error occurred calculating L_latt: {e}")
        L_latt = np.nan


    # Normalize terms - handle inf and nan explicitly
    C_norm = 0
    if not np.isclose(C_latt, C_rand) and not np.isnan(C_rand) and not np.isnan(C_latt):
         C_norm = (C_latt - C_obs) / (C_latt - C_rand)

    L_norm = 0
    if not np.isclose(L_latt, L_rand) and not np.isnan(L_rand) and not np.isnan(L_latt):
        L_norm = (L_obs - L_rand) / (L_latt - L_rand)


    # Final SWP formula
    swp = 1 - np.sqrt(C_norm**2 + L_norm**2)

    # Handle potential issues with swp calculation resulting in non-finite values
    if not np.isfinite(swp):
        print("SWP calculation resulted in a non-finite value. Check graph connectivity and metrics.")
        swp = np.nan # Or handle as appropriate

    return swp, {
        'C_obs': C_obs, 'L_obs': L_obs,
        'C_rand': C_rand, 'L_rand': L_rand,
        'C_latt': C_latt, 'L_latt': L_latt
    }
```
